```python
from sklearn import metrics,datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
wine=datasets.load_wine()
for i in range(1,11):
    model=RandomForestClassifier(n_estimators=i,max_depth=6)
    train_x,test_x,train_y,test_y=train_test_split(wine.data,wine.target,test_size=0.2,random_state=2)
    model.fit(train_x,train_y)
    pred_y=model.predict(test_x)
    print("Number of trees:",i)
    score=model.feature_importances_
    print('Feature importance:')
    for j,k in zip(wine["feature_names"],score):
            print(j,":",k )
    print("Accuracy:",metrics.accuracy_score(test_y,pred_y))
    print('===================')
    
    
```
