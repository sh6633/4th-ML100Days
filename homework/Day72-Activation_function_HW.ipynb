# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(X):
    l=[]
    for i in X:
        if i>=0:
            l.append(i)
        else:
            l.append(0)
    return l
def dReLU(X):
    l=[]
    for i in X:
        if i>=0:
            l.append(i)
        else:
            l.append(0)   
x=np.linspace(-50,100,100)
plt.title("ReLU")
plt.grid()
plt.text(-20,80,"f(x)=max(0,x)",fontsize=20)
#print(ReLU(x))
plt.plot(x,ReLU(x),color="r")
plt.show()
```
