- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
estimators = [("k_means_8",KMeans(n_clusters=8)),
             ("k_means_5",KMeans(n_clusters=3)),
             ('k_means_bad_init',KMeans(n_clusters=5, n_init=2,
                                       init="random")),
             ("k_means_bad_init2",KMeans(n_clusters=5,n_init=1,
                                      init="random"))]
```

```python
print(X[:,0].shape,X[:,1].shape,y.shape)
```

```python
fignum=1
titles = ["k_means_8","k_means_3",'k_means_bad_init',"k_means_bad_init2"]
for name, est in estimators:
    fig =plt.figure(fignum, figsize=(10,10))
    ax = Axes3D(fig, rect=[0,0,.95,1],elev=48, azim=134)
    
    est.fit(X)
    labels=est.labels_
    ax.scatter(X[:,0],X[:,1]
               ,y,c=labels.astype(np.float), edgecolor="k")
    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_title(titles[fignum-1])
    ax.dist = 12
    fignum = fignum + 1
    
```
